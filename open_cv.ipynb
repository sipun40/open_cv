```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
from google.colab import files
uploaded = files.upload()
```

```python
filename=list(uploaded.keys())[0]
img=cv2.imread(filename)
cv2_imshow(img)
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray)
```

```python
resized = cv2.resize(img,(300,300))
cv2_imshow(resized)
```

```python
a =[img,gray]
for i in a:
  resized = cv2.resize(i,(300,300))
  cv2_imshow(resized)
```

```python
annotated = img.copy()
cv2.rectangle(annotated, (50, 50), (250, 250), (0,0,255), 2)
cv2.putText(annotated,'My Pc',(55,45),cv2.FONT_HERSHEY_SIMPLEX,1,(0,0,255),2)
cv2_imshow(annotated)
```

```python
cv2.imwrite('output_image.jpg',annotated)
print("img saved as output_image.jpg")
```

```python
!pip install tensorflow
import matplotlib.pyplot as plt

from tensorflow.keras.utils import load_img
from tensorflow.keras.utils import img_to_array
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
uploaded = files.upload()
filename = list(uploaded.keys())[0]
```

```python
img = load_img(filename,target_size=(224,224))
img_array = img_to_array(img)
img_array = np.expand_dims(img_array,axis=0)

```

```python
datagen = ImageDataGenerator(
    rotation_range=30,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
    brightness_range=[0.5,1.5]
)
```

```python
i=0
for batch in datagen.flow(img_array,batch_size=1):
  plt.figure(figsize=(2,2))
  plt.imshow(batch[0].astype('uint8'))
  plt.axis('off')
  i+=1
  if i>5:
    break
```
